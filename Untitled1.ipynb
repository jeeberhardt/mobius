# Rosetta MHC

```python
url = 'https://rosettamhc.chemistry.ucsc.edu'
json_url = 'https://rosettamhc.chemistry.ucsc.edu/static/%s.json'
pdb_url = 'https://rosettamhc.chemistry.ucsc.edu/static/%s-ensemble.pdb'
```

```python
http = urllib3.PoolManager(cert_reqs = 'CERT_NONE')
resp = http.request("GET", url=url)
soup = BeautifulSoup(resp.data, 'html.parser')
peptides = [o.text for o in soup.find_all('option')[1:]]
```

```python
! mkdir 'data'
! mkdir 'data/rosetta_mhc'
! mkdir 'data/rosetta_mhc/pdb'
```

```python
data = []

for peptide in peptides:
    print(peptide)

    # Download PDB file
    resp = http.request("GET", url=pdb_url % peptide)
    with open('data/rosetta_mhc/pdb/%s-ensemble.pdb' % peptide, 'w') as w:
        w.write(resp.data.decode("utf-8"))

    # Get JSON file
    resp = http.request("GET", url=json_url % peptide)
    json_data = json.loads(resp.data)

    data.append((peptide, len(peptide), float(json_data['ba']),
                 float(json_data['energy']), json_data['template']))

columns = ['peptide', 'length', 'binding_affinity', 'energy', 'template']
df = pd.DataFrame(data=data, columns=columns)
df.to_csv('data/rosetta_mhc/peptide_binding_affinity.csv', index=False)
```

## Correlation using all MHC structures

```python
peptides_selection = df[df['length'] == 9].sort_values(by='binding_affinity')

solvent_exposure = []

for peptide in peptides_selection.itertuples():
    print(peptide.peptide)

    structure = freesasa.Structure('data/rosetta_mhc/pdb/%s-ensemble.pdb' % peptide.peptide)
    result = freesasa.calc(structure)
    
    surface_contact = [result.residueAreas()['A']['%s' % i].relativeSideChain for i in range(181, 190)]
    solvent_exposure.append(surface_contact)

solvent_exposure = np.array(solvent_exposure)
solvent_exposure = np.nan_to_num(solvent_exposure, nan=0)
```

```python
np.mean(solvent_exposure, axis=0)
```

```python
np.std(solvent_exposure, axis=0)
```

```python
ff = ForceField('mobius/data/parameters.csv')
ff._term_parameters['vdw']['score_max'] = 5
ff._term_parameters['vdw']['smooth'] = 0.2
ff._term_parameters['vdw']['n'] = 4
print(ff._term_parameters)

vt = VirtualTarget(ff)
```

```python
target_sequence = peptides_selection.iloc[0]['peptide']
vt.generate_pharmacophore_from_target_sequence(target_sequence, solvent_exposures=solvent_exposure[0])
```

```python
scores = vt.score_peptides(peptides_selection['peptide'], noise=0)
terms = np.array([ff.score(vt.pharmacophore(), p, details=True)[1] for p in peptides_selection['peptide']])
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
sns.regplot(x=peptides_selection['binding_affinity'], y=scores, ax=ax)
ax.plot()
```

```python
def fun(x, binding_affinity, terms):
    scores = np.sum(np.prod((terms, x), axis=0), axis=1)
    return mean_squared_error(binding_affinity, scores)

bnds = ((1, 1000), (1, 1000), (1, 1000), (1, 1000))
res = optimize.minimize(fun, (1, 1, 1, 1), method='SLSQP', args=(peptides_selection['binding_affinity'], terms), bounds=bnds, tol=1e-10)
print('vdW: %f - Electrostatic: %f - HBond: %f - Desolvation: %f' % (res.x[0], res.x[1], res.x[2], res.x[3]))

fig, ax = plt.subplots(figsize=(10, 10))
sns.regplot(x=peptides_selection['binding_affinity'], y=np.sum(np.prod((terms, res.x), axis=0), axis=1), ax=ax)
plt.plot()
```

## Correlation using a subset of structures

```python
resids = []

for peptide in peptides_selection.itertuples():
    print(peptide.peptide)

    u = Universe('data/rosetta_mhc/pdb/%s-ensemble.pdb' % peptide.peptide)
    
    selection = u.select_atoms('byres (around 4 resid 181-190)')
    resids.extend(selection.residues.resids)
    
resids = list(set(resids))
```

```python
peptides = list(peptides_selection['peptide'].values)

# Remove problematic structures
peptides.pop(peptides.index('IIQFPNTYL'))
peptides.pop(peptides.index('RITGLYPTL'))

selection_str = ' or '.join(['(resid %s and not (backbone or name H*))' % i for i in [70, 97, 116, 156, 159]])

mat = np.zeros(shape=(len(peptides), len(peptides)))

for i in range(len(peptides)):
    print(peptides[i])
    u1 = Universe('data/rosetta_mhc/pdb/%s-ensemble.pdb' % peptides[i])
    s1 = u1.select_atoms(selection_str)
    
    for j in range(i + 1, len(peptides)):
        u2 = Universe('data/rosetta_mhc/pdb/%s-ensemble.pdb' % peptides[j])
        s2 = u2.select_atoms(selection_str)
        
        try:
            mat[i, j] = rms.rmsd(s1.positions, s2.positions, superposition=True)
            mat[j, i] = mat[i, j]
        except:
            print(p1, p2)

np.savetxt('data/rosetta_mhc/mat_distance.txt', mat)
```

```python
mat = np.loadtxt('data/rosetta_mhc/mat_distance.txt')

m = MDS(n_init=10, max_iter=1000, n_components=2, dissimilarity='precomputed', n_jobs=-1)
embedding = m.fit_transform(mat)
print(m.stress_)
```
